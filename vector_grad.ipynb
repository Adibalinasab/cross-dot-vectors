```python
import numpy as np
import matplotlib.pyplot as plt

def get_vectore_input(prompt):
  while True:
    user_input=input(prompt)
    parts=user_input.replace(',','').split()
    if len(parts)==3:
       try:
           return np.array([float(p)for p in parts])
       except ValueError:
             print("just enter numbers")
    else:
         print("enter 3 numbers")
def plot_vectors(v1,v2,cross):
 fig=plt.figure(figsize=(8,8))
 ax=fig.add_subplot(111,projection='3d')
 origin=[0,0,0]
 ax.quiver(*origin,*v1,color='blue',label='vector A',arrow_length_ratio=0.1)
 ax.quiver(*origin,*v2,color='green',label='vectorB',arrow_length_ratio=0.1)
 ax.quiver(*origin,*cross,color='red',label='AcrossB',arrow_length_ratio=0.1)
 limit=max(np.max(np.abs(v1)),np.max(np.abs(v2)),np.max(np.abs(cross)),1)
 ax.set_xlim([-limit,limit])
 ax.set_ylim([-limit,limit])
 ax.set_zlim([-limit,limit])
 ax.set_xlabel('x')
 ax.set_ylabel('y')
 ax.set_zlabel('z')
 ax.set_title('vector visualization')
 ax.legend()
 plt.show()
def main():
  v1=get_vectore_input("input vector number one like (x  y  z ):")
  v2=get_vectore_input("input vector number two like (x y z)")
  dot=np.dot(v1,v2)
  cross=np.cross(v1,v2)
  print("-"*20)
  print(f"vector 1:{v1}")
  print(f"vector 2 :{v2}")
  print(f"dot product:{dot}")
  print(f"cross product:{cross}")
  plot_vectors(v1,v2,cross)
main()

```
